```python
import torch
import torch.optim
import torch.nn as nn
import torch.nn.functional as F
```

```python
loss_func = nn.BCELoss()
y = torch.tensor([0,1,0], dtype=torch.float)
y_pred = torch.tensor([0,0.3,0], dtype=torch.float)
print(loss_func(y_pred, y).numpy())
```

```python
import matplotlib.pyplot as plt
import numpy as np
def asd(x):
    y = torch.tensor([0,1,0], dtype=torch.float)
    y_pred = torch.tensor([0,x,0], dtype=torch.float)
    return loss_func(y_pred, y).numpy()
x = np.arange(0,1,0.1)
y = [asd(x) for x in x]
plt.plot(x,y)
```
